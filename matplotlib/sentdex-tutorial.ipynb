```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [1,2,3]
y1 = [4,5,6]

x2 = np.arange(1, 4, 1)
y2 = np.arange(10, 16, 2)

plt.plot(x1, y1, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.title('Interesting Graph\n')
```

```python
plt.show()
```

## Tutorial 3 - Bar Chart and Histogram

```python
x1 = [2, 4, 6, 8, 10]
y1 = [6, 7, 8, 2, 4]

# Notice that the point (2,6) from the x1-y1 list will not show up, hiding behind (2, 22) from x2-y2
x2 = [2, 3, 5, 9, 11]
y2 = [22, 1, 9, 7, 4]

plt.bar(x1, y1, label='Bars1')
plt.bar(x2, y2, label='Bars2')
plt.legend()
```

```python
plt.show()
```

```python
population_ages = np.random.randint(1, high=95, size=50)
ids = [x for x in range(len(population_ages))]
plt.bar(ids, population_ages)
population_ages
```

```python
plt.show() # this will only show who's the oldest, not the distribution
```

```python
bins = np.arange(-20, 110, 10)
bins
```

```python
plt.hist(population_ages, bins, histtype='bar', rwidth=0.9)
plt.show()
```

## Tutorial 4 - Scatter Plots

```python
x = np.linspace(0, 100)
y = np.random.random(50) * 100

plt.scatter(x, y, marker='X', s=100)
plt.show()
```

## Tutorial 5 - Stack Plots

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 6, 5.5, 8, 10]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 5, 5, 2]
playing = [5, 3, 7, 4, 9]
labels = ['sleeping', 'eating', 'working', 'playing']
colors = plt.get_cmap('tab20').colors
plt.stackplot(days, sleeping, eating, working, playing, labels=labels, colors=colors)
plt.title('Week 1 Activities')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend()
plt.show()
```

## Tutorial 6 - Pie Charts

```python
slices = [7, 2, 2, 13]
plt.pie(slices, colors=colors[4:9], shadow=True, startangle=90, explode=[0.05, 0.05, 0.1, 0.05], labeldistance=0.5, autopct='%1.1f%%')
plt.legend(labels=labels)
plt.show()
```
